```python
import numpy as np
import pandas as pd
import pickle
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
movies=movies.merge(credits,on='title')
#merging both files on the basis of title
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
# fetching cast name
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
# fetching director name 
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
#converting string into lists
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])

# we are removing space between the names
```

```python
movies.head()
```

```python
movies['fulldata'] = movies['overview'] + movies['genres'] +movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
newmovies = movies[['movie_id' , 'title','fulldata']]
```

```python
newmovies
```

```python
newmovies['fulldata'] = newmovies['fulldata'].apply(lambda x:" ".join(x))
```

```python
newmovies.head()
```

```python
newmovies['fulldata'] = newmovies['fulldata'].apply(lambda x:x.lower())
```

```python
newmovies.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
newmovies['fulldata'] = newmovies['fulldata'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000,stop_words='english')
```

```python
vectors = cv.fit_transform(newmovies['fulldata']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
# checking similarity of each movie with each other
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
# similarity  1 with 1 , 1 with 2 , 1 with 3 and so on.
```

```python
def recomendation(movie):
    movie_index = newmovies[newmovies['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(newmovies.iloc[i[0]].title)
```

```python
recomendation('Spectre')
```

```python
def filterUsingGenre(genre):
    movielist=[]
    for index,row in movies.iterrows():
        if genre in row['genres']:
            movielist.append(row['title'])
    return movielist[:10]
```

```python
import cv2
import numpy as np
from keras.models import model_from_json 
import warnings
warnings.filterwarnings("ignore")


emotion_dict = {0: "Angry", 1: "Disgusted", 2: "Fearful", 3: "Happy", 4: "Neutral", 5: "Sad", 6: "Surprised",7:"No Face"}
json_file = open('C:/Users/admin/Desktop/Face_Recognition/emotion_model.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
emotion_model = model_from_json(loaded_model_json)

# load weights into new model
emotion_model.load_weights("C:/Users/admin/Desktop/Face_Recognition/emotion_model.h5")
print("Loaded model from disk")

# start the webcam feed
cap = cv2.VideoCapture(0)

while True:
  
    ret, frame = cap.read()
    frame = cv2.resize(frame, (1280, 720))
    if not ret:
        break
    face_detector = cv2.CascadeClassifier('C:/Users/admin/anaconda3/Lib/site-packages/cv2/data/haarcascade_frontalface_default.xml')
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # detect faces available on camera
    num_faces = face_detector.detectMultiScale(gray_frame, scaleFactor=1.3, minNeighbors=5)

    # take each face available on the camera and Preprocess it
    for (x, y, w, h) in num_faces:
        cv2.rectangle(frame, (x, y-50), (x+w, y+h+10), (0, 255, 0), 4)
        roi_gray_frame = gray_frame[y:y + h, x:x + w]
        cropped_img = np.expand_dims(np.expand_dims(cv2.resize(roi_gray_frame, (48, 48)), -1), 0)

        # predict the emotions
        if cv2.waitKey(1) & 0xFF == ord('c'):
            emotion_prediction = emotion_model.predict(cropped_img)
            maxindex = int(np.argmax(emotion_prediction))
            cv2.putText(frame, emotion_dict[maxindex], (x+5, y-20), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_AA)
            cv2.imshow('Emotion Detection', frame)
            print(emotion_dict[maxindex])
            if emotion_dict[maxindex] == 'Happy':
                arr=filterUsingGenre('Comedy')
                print(np.array(arr[:5]))
            elif emotion_dict[maxindex] == 'Neutral':
                arr=filterUsingGenre('Action')
                print(np.array(arr[:5]))
            elif emotion_dict[maxindex] == 'Angry':
                arr=filterUsingGenre('Fantasy')
                print(np.array(arr[:5]))
            elif emotion_dict[maxindex] =='sad':
                arr = filterUsingGenre('Adventure')
                print(np.array(arr[:5]))
            elif emotion_dict[maxindex] =='angry':
                arr = filterUsingGenre('Thriller')
                print(np.array(arr[:5]))
            elif emotion_dict[maxindex] =='Surprised':
                arr = filterUsingGenre('Crime')
                print(np.array(arr[:5]))
        cv2.putText(frame, emotion_dict[maxindex], (x+5, y-20), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2, cv2.LINE_AA)
        cv2.imshow('Emotion Detection', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
